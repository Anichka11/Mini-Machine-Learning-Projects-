```python
import librosa
signal,fs = librosa.load("guitar.mp3")
```

```python
from IPython.display import Audio, display
display(Audio(signal, rate=fs,autoplay=True))
```

```python
from scipy.signal import periodogram
import matplotlib.pyplot as plt
f,pxx = periodogram(x=signal,fs=fs)
plt.plot(f,pxx)
```

```python
librosa.display.waveshow(y=signal,sr=fs)
```
