```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
mymodel = KMeans(n_clusters=4,max_iter=45000)
X,_ =make_blobs(n_samples=3000,n_features=2,centers=[[10,14],[13,30],[18,19],[24,30]],random_state=1)
mymodel.fit(X)
colors = mymodel.predict(X)
plt.scatter(X[:,0],X[:,1],c=colors)
plt.scatter(mymodel.cluster_centers_[:,0],mymodel.cluster_centers_[:,1],c='pink',s=90)
plt.show()
```
