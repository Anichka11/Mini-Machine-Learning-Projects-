```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")
data.drop("PassengerId",axis=1,inplace=True)
data.head()
```

```python
data['Name'] =data['Name'].map(lambda test:test.split(",")[1].split(".")[0].strip())
data.head()
```

```python
data['Age'] = data.groupby(['Survived', 'Pclass', 'Name'])['Age'].transform(lambda x:x.fillna(x.mean()))
data["Embarked"] = data['Embarked'].fillna("S")
data.isnull().sum()
# data.groupby(['Survived', 'Pclass', 'Name'])['Age'].mean()
```

```python
import re
def get_deck_number(text):
  if type(x)==float:
    return 0
  result = re.findall('d\+',x)
  if len(result)==0:
    return 0
  else:
    return float(result[0])
data['Deck']=data['Cabin'].map(get_deck_number)
data.head()
# for x in data['Cabin'].unique():
#   if type(x)==float:
#     print(0)
#   else:
#     result = re.findall('d\+',x)
#     if len(result)==0:
#       print(0)
#     else:

#       print(float(result[0]))
```

```python
def get_deck_letter(x):
  if type(x)==float:
    return 'C'
  else:
    result =re.findall(r'[a-zA-Z]',x)
    text ="".join(result)
    if len(text)==1:
      return text
    elif len(text)>=2 and text[0]==text[1]:
      return text[0]
    else:
      return text
    if len(text)>2:
      return text[0]
data['deck_letter'] =data['Cabin'].map(get_deck_letter)
data.head()
```

```python
def get_martial_status(Name):
  if Name in ['Miss',"Mrs","Master","Don"]:
    return  "Married"
  else:
    return "Unknown"
data['Name'] =data['Name'].map(get_martial_status)
data['Name'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for column in data.columns:
  if data[column].dtype=='object':
    data[column]=label.fit_transform(data[column])
data.head()
```

```python
data["Family Size"]=data['SibSp']+data['Parch']
data.drop(['SibSp','Parch','Sex'], axis=1, inplace=True)
data.head()
```
